# 2. Standardizing contact details in an Excel file for a marketing campaign using  customer data

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('us_customer_data 1.csv')
```

```python
countries=pd.read_csv(r'Country-codes.csv',encoding='latin1')
```

```python
countries
```

```python
def country(df):
    for index,row in df.iterrows():
        df.at[index,'Country_Code']=row['address'][-8:-6]
```

```python
#df['Country_Code']=df['Country_Code'].str.slice(-8,-6)
```

```python
country(df)
```

```python
df['phone']=df['phone'].str.replace('-','')
```

```python
df.head(2)
```

```python
# merge_df=pd.merge('df','countries',on='country_code',how='left')
```

```python
df['phone']=df['phone'].astype(str)
for i in range(len(df)):
    code = df.loc[i,'Country_Code']
    found=False
    for j in range(len(countries)):
        if code == countries.loc[j,'Country_code']:
            dial = countries.loc[j,'International_dialing']
            df.loc[i,'phone']=(str(dial)+" "+df.loc[i,'phone'])
            found = True
            break
    if not found:
        df.loc[i,'phone']=('+1 '+df.loc[i,'phone'])
        
```
