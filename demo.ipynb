### Assigning the condition to the amount fields 

```python
if merge_df['transaction_id'].isnull().any:
    merge_df['amount']=merge_df['amount'].fillna(0)
else: 
    mean_amount=merge_df[merge_df['amount'].mean()]
    merge_df['amount']=merge_df['amount'].fillna(mean_amount)
```
